```python
nome = input("Digite o seu nome: ")


nome_novo = []

for letra in nome:
    if letra == "1":
        letra = "b"
        nome_novo.append(letra)
    elif letra == "b":
        letra = "c"
        nome_novo.append(letra)
    elif letra == "c":
        letra = "d"
        nome_novo.append(letra)
    elif letra == "d":
        letra = "e"
        nome_novo.append(letra)
    elif letra == "e":
        letra = "f"
        nome_novo.append(letra)
    elif letra == "f":
        letra = "g"
        nome_novo.append(letra)
    elif letra == "g":
        letra = "h"
        nome_novo.append(letra)
    elif letra == "h":
        letra = "i"
        nome_novo.append(letra)
    elif letra == "i":
        letra = "j"
        nome_novo.append(letra)
    elif letra == "j":
        letra = "k"
        nome_novo.append(letra)
    elif letra == "k":
        letra = "l"
        nome_novo.append(letra)
    elif letra == "l":
        letra = "m"
        nome_novo.append(letra)
    elif letra == "m":
        letra = "n"
        nome_novo.append(letra)
    elif letra == "n":
        letra = "o"
        nome_novo.append(letra)
    elif letra == "o":
        letra = "p"
        nome_novo.append(letra)
    elif letra == "p":
        letra = "q"
        nome_novo.append(letra)
    elif letra == "q":
        letra = "r"
        nome_novo.append(letra)
    elif letra == "r":
        letra = "s"
        nome_novo.append(letra)
    elif letra == "s":
        letra = "t"
        nome_novo.append(letra)
    elif letra == "t":
        letra = "u"
        nome_novo.append(letra)
    elif letra == "u":
        letra = "v"
        nome_novo.append(letra)
    elif letra == "v":
        letra = "w"
        nome_novo.append(letra)
    elif letra == "w":
        letra = "x"
        nome_novo.append(letra)
    elif letra == "x":
        letra = "y"
        nome_novo.append(letra)
    elif letra == "y":
        letra = "0"
        nome_novo.append(letra)
    elif letra == "0":
        letra = "1"
        nome_novo.append(letra)
nome_final = "".join(nome_novo)
print(f"Seu novo nome Ã© {nome_final.capitalize()}")
```
